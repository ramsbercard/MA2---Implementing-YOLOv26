```python
import torch
import os
import pandas as pd
from ultralytics import YOLO

print("Ultralytics Version:", YOLO.__module__)
print("CUDA Available:", torch.cuda.is_available())
print("GPU:", torch.cuda.get_device_name(0))
print("Capability:", torch.cuda.get_device_capability(0))
print("Torch CUDA:", torch.version.cuda)

device = 0 if torch.cuda.is_available() else "cpu"
print("Using device:", device)
```

```python


DATA_PATH = os.path.abspath("VehicleDataset/images/data.yaml")

if os.path.exists(DATA_PATH):
    print("Dataset YAML found ✅")
else:
    print("Dataset YAML NOT found ❌")
```

```python
model1 = YOLO("yolo26n.pt")

results1 = model1.train(
    data=DATA_PATH,
    epochs=25,
    imgsz=640,
    optimizer="AdamW",
    batch=4,
    lr0=0.01,
    device="0",
    project="/Ramil_machine_learning/yolo26_runs",
    name="model1_25ep"
)
```

```python
import multiprocessing
multiprocessing.set_start_method("spawn", force=True)

model2 = YOLO("yolo26s.pt")

results2 = model2.train(
    data=DATA_PATH,
    epochs=30,
    imgsz=640,
    optimizer="SGD",
    batch=20,
    lr0=0.001,
    workers =2,
    device="0",
    project="/Ramil_machine_learning/yolo26_runs",
    name="model2_30ep"
)
```

```python
import multiprocessing
multiprocessing.set_start_method("spawn", force=True)

model3 = YOLO("yolo26m.pt")

results3 = model3.train(
    data=DATA_PATH,
    epochs=40,
    imgsz=640,
    optimizer="auto",
    batch=-1,   # Auto batch
    lr0=0.0001,
    workers =2,
    device="0",
    project="/Ramil_machine_learning/yolo26_runs",
    name="model3_40ep"
)
```

```python
from IPython.display import Image, display
import os

def show_results(run_name):
    base_path = f"/Ramil_machine_learning/yolo26_runs/{run_name}"
    
    results_img = os.path.join(base_path, "results.png")
    confusion_img = os.path.join(base_path, "confusion_matrix.png")
    
    if os.path.exists(results_img):
        display(Image(results_img))
        
    if os.path.exists(confusion_img):
        display(Image(confusion_img))

# Display all
show_results("model1_25ep4")
show_results("model2_30ep")
show_results("model3_40ep2")
```

```python
model1_best = YOLO("/Ramil_machine_learning/yolo26_runs/model1_25ep4/weights/best.pt")
model2_best = YOLO("/Ramil_machine_learning/yolo26_runs/model2_30ep/weights/best.pt")
model3_best = YOLO("/Ramil_machine_learning//yolo26_runs/model3_40ep2/weights/best.pt")

metrics1 = model1_best.val()
metrics2 = model2_best.val()
metrics3 = model3_best.val()
```

```python
def compute_f1(p, r):
    return 2*(p*r)/(p+r) if (p+r) != 0 else 0

performance_table = pd.DataFrame({
    "Model": ["Model1_25ep", "Model2_30ep", "Model3_40ep"],
    "mAP50": [
        metrics1.box.map50,
        metrics2.box.map50,
        metrics3.box.map50
    ],
    "Precision": [
        metrics1.box.mp,
        metrics2.box.mp,
        metrics3.box.mp
    ],
    "Recall": [
        metrics1.box.mr,
        metrics2.box.mr,
        metrics3.box.mr
    ],
    "F1 Score": [
        compute_f1(metrics1.box.mp, metrics1.box.mr),
        compute_f1(metrics2.box.mp, metrics2.box.mr),
        compute_f1(metrics3.box.mp, metrics3.box.mr)
    ]
})

performance_table
```
